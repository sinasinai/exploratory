```python
gdp = manufacturing.merge(retail,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(construction,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(affh,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(pbs,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(tw,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(mining,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(firl,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(wt,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(aeraf,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(oseg,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(eshs,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(info,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(utilities,how='inner',left_index=True,right_index=True)

gdp = gdp.merge(rer,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(services,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(goods,how='inner',left_index=True,right_index=True)
gdp = gdp.merge(real_gdp,how='inner',left_index=True,right_index=True)
```
