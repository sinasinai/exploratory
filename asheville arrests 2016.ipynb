```python
offense = avl.groupby('OFFENSE').sum()[cols].reset_index()
```

```python
offense['offense_value'] = offense.OFFENSE.map(offense_dict)
```

```python
offense['pct_adult_black_arrests'] = offense.AB/offense.AB.sum()
offense['pct_adult_white_arrests'] = offense.AW/offense.AW.sum()
```

```python
offense['black_to_white_ratio'] = offense.pct_adult_black_arrests/offense.pct_adult_white_arrests
```

```python
offense.sort_values(by='black_to_white_ratio',ascending=False)
```

```python
offense['pct_white'] = offense.AW/(offense.AW + offense.AB + offense.AI + offense.AA + offense.AH + offense.AN)
offense['pct_black'] = offense.AB/(offense.AW + offense.AB + offense.AI + offense.AA + offense.AH + offense.AN)
```

```python
offense.sort_values(by='pct_black',ascending=False)
```
