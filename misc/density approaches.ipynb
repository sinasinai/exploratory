```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import math
```

```python
math.cos(np.pi*2)
```

```python
df = pd.DataFrame(np.random.uniform(0,1,size=(1000,1)),columns=['r'])
df['theta'] = np.random.uniform(0,2*np.pi,size=(1000,1))
```

```python
df['x'] = df.apply(lambda row: row['r'] * math.cos(row['theta']),axis=1)
df['y'] = df.apply(lambda row: row['r'] * math.sin(row['theta']),axis=1)
```

```python
sns.scatterplot(x = df.x,y=df.y)
plt.show()
```

```python
np.sqrt(df.x.std() + df.y.std())
```

```python
np.sqrt(df.x.std()*df.y.std())
```
