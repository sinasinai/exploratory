# Parcel Data

Can also query parcel data using their API here https://data.buncombecounty.org/datasets/buncombe-county-parcels

```python
parcels = pd.read_csv('~/Desktop/Buncombe_County_Parcels.csv',low_memory=False)
```

```python
cols = ['TotalMarketValue','AppraisedValue','TaxValue','LandValue','BuildingValue','SalePrice','Acreage']
owners = parcels.loc[parcels['CityName'] == 'ASHEVILLE'].groupby('Owner').sum().sort_values('TotalMarketValue',ascending=False)[cols]
```

```python
tmv = sns.histplot(owners['TotalMarketValue'].sample(1000))
tmv.set_title('Asheville Property Owner Market Value Distribution - Sample')
plt.show()
```

```python
acr = sns.histplot(owners['Acreage'].sample(1000))
acr.set_title('Asheville Property Owner Acreage Distribution - Sample')
plt.show()
```
